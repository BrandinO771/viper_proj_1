```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
ratings = pd.read_csv("API Exported Data.csv")
c_s = ratings.loc[ratings["Function"] == "Consumer Satisfaction"]

valueofcare = pd.read_csv("Payment_and_value_of_care_-_Hospital_CLEAN.csv")

dbdf = pd.read_csv("diabetes_only_2017_chrgs.csv")

zips = pd.read_csv("All Zip Data.csv")

dbdf = pd.merge(dbdf, c_s, on="Provider ID")
dbdf = pd.merge(dbdf, zips, on="Zipcode")

dbdf.to_csv("dbdf.csv")
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(20)
fig.set_figheight(10)
slope, inter, rvalue, pvalue, std = stats.linregress(dbdf["Rank Percentile"], dbdf["Average Total Payments"])
yvals = np.linspace(dbdf["Rank Percentile"].min(), dbdf["Rank Percentile"].max())
fit = slope * yvals + inter
ax.set_ylim(0, 30000)
ax.set_xlim(0, 1)
ax.plot(yvals, fit, color="green", linewidth=2, linestyle="--")
ax.scatter(dbdf["Rank Percentile"], dbdf["Average Total Payments"])
ax.set_xlabel("Hospital Consumer Satisfaction Rank Percentile (Best Ranking Hospitals = 1)", fontsize=20)
ax.set_ylabel("Cost of Diabetes Procedure: '638 - DIABETES WCC'", fontsize=20)
plt.legend(["Regression Line", "Hospitals"])
pvalue = '{0:.30f}'.format(pvalue)
plt.title("Cost of Procedure: '638 DIABETES WWC'\nvs.\nHospital Consumer Satisfaction Rank Percentile (National)", fontsize=20)
plt.text(.01, 26000, "Pvalue: " + str(pvalue) + "\nRvalue: " + str(rvalue) + "\nRsquared: -" + str(rvalue**2), fontsize=16)
plt.savefig("Rank Percentile vs Diabeties cost.png")
```
