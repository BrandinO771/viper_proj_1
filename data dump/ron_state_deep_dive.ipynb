```python
# 871	Septicemia      
# 470	major joint replace
# 291	Heart Failure 
# 392	GI / Digestive disorders   
# 638	638 - DIABETES W CC
search_1_item = '291 - HEART FAILURE & SHOCK W MCC'
search_2_item = '392 - ESOPHAGITIS, GASTROENT & MISC DIGEST DISORDERS W/O MCC'
search_3_item = '470 - MAJOR JOINT REPLACEMENT OR REATTACHMENT OF LOWER EXTREMITY W/O MCC'
search_4_item = '638 - DIABETES W CC'
search_5_item = '871 - SEPTICEMIA OR SEVERE SEPSIS W/O MV 96+ HOURS W MCC'

df_a = pd.read_csv("Inpatient_PYMT_IPPS_DRG_FY2017.csv")
df_291_ = df_a.loc[df_a["DRG Definition"] == search_1_item,  :]
df_392_ = df_a.loc[df_a["DRG Definition"] == search_2_item,  :]
df_470_ = df_a.loc[df_a["DRG Definition"] == search_3_item,  :]
df_638_ = df_a.loc[df_a["DRG Definition"] == search_4_item,  :]
df_871_ = df_a.loc[df_a["DRG Definition"] == search_5_item,  :]

result_df = pd.concat([df_291_, df_392_, df_470_, df_638_, df_871_ ], ignore_index=True)
#result_df.head(10)
result_df.tail(10)
```

```python
result_df.to_csv("Inpatient_PYMT_IPPS_DRG_FY2017_CLEAN.csv", index=False)
```
