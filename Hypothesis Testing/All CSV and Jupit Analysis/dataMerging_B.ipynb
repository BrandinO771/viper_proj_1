```python
#reformat Census with buckets 
import pandas as pd
import matplotlib.pyplot as plt

df_cen = pd.read_csv("All USA Census by Zip Data.csv")
income_cat = []
income_cat_num = []
row = 1         

for row in range(len(df_cen['median Household Income'])) : 
    if  df_cen['median Household Income'][row] < 45000 :
        income_cat.append('low income')
        income_cat_num.append(1)
    if  df_cen['median Household Income'][row] >= 45000 and  df_cen['median Household Income'][row]< 140000 : 
        income_cat.append('middle income')
        income_cat_num.append(2)        
    if  df_cen['median Household Income'][row] >= 140000 :         
        income_cat.append('high income') 
        income_cat_num.append(3)        


region_type = []
region_type_num = []
rowz = 1 
for rowz in range( len(df_cen['Population Density']) ) : 

    if ( df_cen['Population Density'][rowz]>=3000 and    df_cen['Population'][rowz]>= 50000 )  :
        region_type.append('urban')
        region_type_num.append(3)        
    if ( df_cen['Population Density'][rowz]>=3000 and    df_cen['Population'][rowz]< 50000 )  :
        region_type.append('suburban')        
        region_type_num.append(2) 
    if ( df_cen['Population Density'][rowz]>=1000  and    df_cen['Population Density'][rowz]<3000)   and  df_cen['Population'][rowz]>= 0   :
        region_type.append('suburban')   
        region_type_num.append(2)         
    if ( df_cen['Population Density'][rowz]>=0      and    df_cen['Population Density'][rowz]<1000     and  df_cen['Population'][rowz]>=0 )  :
        region_type.append('rural')   
        region_type_num.append(1)  
    if  ( df_cen['Population Density'][rowz]<0   and   df_cen['Population'][rowz]<0 )  :
        region_type.append('deserted')
        region_type_num.append(0) 
     
```

```python
print(len( df_cen['Population']))
print(len(df_cen['Population Density']))
print(len(region_type))
```

```python

df_cen['region cat'] = region_type
df_cen['region cat num'] = region_type_num 
df_cen['income cat'] = income_cat 
df_cen['income cat num']= income_cat_num 
df_cen.to_csv("All USA Census by Zip Data Cats_NamsNums.csv", index=False)
df_cen.head()  
```
