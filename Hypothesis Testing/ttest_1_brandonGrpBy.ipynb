```python

# Dependencies
import numpy as np
import pandas as pd
from scipy import stats
# assign file name 
fil1 = "JOINT_BY_STATE__NY - Manhattan_brandon_.csv" 
fil2 = "JOINT_BY_STATE__NY - East Long Island_brandon_.csv" 
#read in our file
sample1 = pd.read_csv(fil2)
sample2 = pd.read_csv(fil3)
# Create two samples
s1 = sample1['Average Covered Charges']
s2 = sample2['Average Covered Charges']
# Run T test
(t_stat, p) = stats.ttest_ind(s1, s2, equal_var=False)
print("t-statistics is {}.".format(t_stat))
print("p-value is {}.".format(p))
```
